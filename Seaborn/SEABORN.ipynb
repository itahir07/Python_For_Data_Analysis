# Histrogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230])
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True)
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True,rug=True)
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"],log_scale=True)
plt.show()
```
